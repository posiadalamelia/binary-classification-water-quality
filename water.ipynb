### Przedstawienie dostępnych zmiennych

#### Cechy wejściowe
<b>PH wody:</b><br>
Miara kwasowości lub zasadowości, która jest istotna dla zachowania jakości wody pitnej i zapobiegania problemom zdrowotnym. Światowa Organizacja Zdrowia zaleca maksymalne dopuszczalne pH od 6.5 do 8.5.

<b>Twardość wody (Hardness):</b><br>
Miara zawartości związków mineralnych, głównie soli wapnia i magnezu. Sole te są rozpuszczane ze złóż geologicznych, przez które przepływa woda.

<b>Substancje stałe (Solids - Total dissolved solids - TDS):</b><br>
Woda ma zdolność rozpuszczania szerokiej gamy nieorganicznych i niektórych organicznych minerałów lub soli, takich jak potas, wapń, sód, wodorowęglany, chlorki, magnez, siarczany itp. Jest to ważny parametr dla wykorzystania wody. Woda o wysokiej wartości TDS wskazuje, że woda jest silnie zmineralizowana. Pożądana granica dla TDS wynosi 500 mg/l, a maksymalna granica dla wody przeznaczonej do picia wynosi 1000 mg/L.

<b>Chloraminy (Chloramines):</b><br>
Chlor i chloramina są głównymi środkami dezynfekującymi stosowanymi w publicznych instalacjach wodnych. Chloraminy są najczęściej tworzone, gdy amoniak jest dodawany do chloru w celu leczenia wody pitnej. Poziomy chloru do 4 miligramów na litr (mg/L lub 4 części na milion (ppm)) są uważane za bezpieczne w wodzie pitnej.

<b>Siarczan (Sulfate):</b><br>
Siarczany są naturalnie występującymi substancjami, które znajdują się w minerałach, glebie i skałach. Stężenie siarczanów w wodzie morskiej wynosi około 2700 miligramów na litr (mg/L). W większości wodociągów słodkowodnych mieści się ona w zakresie od 3 do 30 mg/L.

<b>Zdolność przewodzenia prądu (Conductivity):</b><br>
Czysta woda nie jest dobrym przewodnikiem prądu elektrycznego. Wzrost stężenia jonów zwiększa przewodność elektryczną wody. Ilość rozpuszczonych ciał stałych w wodzie określa przewodność elektryczną. Przewodność elektryczna (EC) faktycznie mierzy proces jonowy roztworu, który umożliwia przewodzenie prądu. Zgodnie z normami WHO wartość EC nie powinna przekraczać 400 μS/cm.

<b>Węgiel organiczny (Conductivity):</b><br>
Całkowity węgiel organiczny jest miarą całkowitej ilości węgla w związkach organicznych w czystej wodzie. Zgodnie z amerykańską Agencją Ochrony Środowiska (EPA) < 2 mg/L jako TOC w wodzie uzdatnionej/pitnej, a < 4 mg/L w wodzie źródlanej, która jest używana do uzdatniania.

<b>Trihalometany (Trihalomethanes):</b><br>
THM to substancje chemiczne, które można znaleźć w wodzie uzdatnionej chlorem. Stężenie THM w wodzie pitnej zmienia się w zależności od zawartości materii organicznej w wodzie, ilości chloru wymaganego do uzdatniania wody oraz temperatury uzdatnianej wody. Poziom THM do 80 ppm jest uważany za bezpieczny w wodzie pitnej.

<b>Mętność (Turbidity):</b><br>
Mętność wody zależy od ilości substancji stałych obecnych w stanie zawieszonym. Jest to miara właściwości emitujących światło wody. Zalecana przez WHO wartości to 5,00 NTU.

#### Cecha przewidywana
<b>Zdolność do spożycia (Potability):</b><br>
Wskazuje, czy woda jest bezpieczna do spożycia przez ludzi, gdzie 1 oznacza zdatność do picia, a 0 oznacza niezdatność

#### Statystyki opisowe

```python
# Funkcja do obliczania szczegółowych statystyk opisowych
def detailed_statistics(df):
    stats_dict = {}
    for column in df.select_dtypes(include=[np.number]).columns:
        col_data = df[column].dropna()
        stats_dict[column] = {
            'Średnia': col_data.mean(),
            'Dominanta': col_data.mode().values[0] if not col_data.mode().empty else np.nan,
            'Minimum': col_data.min(),
            'Kwartyl 1': col_data.quantile(0.25),
            'Mediana (Kwartyl 2)': col_data.median(),
            'Kwartyl 3': col_data.quantile(0.75),
            'Maksimum': col_data.max(),
            'Rozstęp': col_data.max() - col_data.min(),
            'IQR': col_data.quantile(0.75) - col_data.quantile(0.25),
            'Odchylenie standardowe': col_data.std(),
            'Wariancja': col_data.var(),
            'Skośność': col_data.skew(),
            'Współczynnik zmienności': col_data.std() / col_data.mean() * 100
        }
    return pd.DataFrame(stats_dict)


statistics = detailed_statistics(df)
display(statistics)

# Zapisanie wyników do pliku CSV
statistics.to_csv('detailed_statistics.csv', encoding='utf-8')
```

#Opis i zastosowanie metod klasyfikacji

### Narządzia używane do analizy wyników modelu


#### Macierz błędu (ang. confusion_matrix)
Macierz błędów jest narzędziem używanym do oceny wyników modelu klasyfikacji binarnej. Pomaga dokładnie zrozumieć, gdzie model popełnia błędy, oraz jak dobrze radzi sobie w klasyfikowaniu różnych kategorii. Zrozumienie tych wskaźników jest kluczowe do oceny i poprawy modelu klasyfikacji jakości wody.

W naszym przypadku model klasyfikuje jakość wody jako zdatną do spożycia lub nie. Na podstawie tego modelu możemy zdefiniować macierz błędów i wyjaśnić jej elementy:
1. Prawdziwie pozytywna klasyfikacja (True Positive, TP): <br> Woda, która jest rzeczywiście zdatna do spożycia, została zaklasyfikowana przez model jako zdatna do spożycia. Przykład: Woda spełniająca wszystkie standardy jakości i oznaczona przez model jako zdatna do spożycia.
2. Prawdziwie negatywna klasyfikacja (True Negative, TN): <br> Woda, która jest rzeczywiście niezdatna do spożycia, została zaklasyfikowana przez model jako niezdatna do spożycia. Przykład: Woda zanieczyszczona substancjami szkodliwymi i oznaczona przez model jako niezdatna do spożycia.
3. Fałszywie negatywna klasyfikacja (False Negative, FN): <br> Woda, która jest rzeczywiście zdatna do spożycia, została zaklasyfikowana przez model jako niezdatna do spożycia. Przykład: Woda spełniająca standardy jakości, ale model błędnie uznał ją za niezdatną do spożycia.
4. Fałszywie pozytywna klasyfikacja (False Positive, FP):<br> Woda, która jest rzeczywiście niezdatna do spożycia, została zaklasyfikowana przez model jako zdatna do spożycia. Przykład: Woda zawierająca szkodliwe zanieczyszczenia, ale model błędnie uznał ją za zdatną do spożycia.

#### Raport z klasyfikacji (ang. classification report)

Za pomocą tych definicji możemy teraz zrozumieć różne wskaźniki używane do oceny modelu:
1. Dokładność (Accuracy): Proporcja wszystkich prawidłowych klasyfikacji (TP i TN) w stosunku do wszystkich przypadków.
$$
{Accuracy}= \frac{TP + TN}{TP + TN + FP + FN}
$$
<br>
2. Czułość (Recall) lub TPR (True Positive Rate): Proporcja prawidłowo zidentyfikowanych pozytywnych przypadków (TP) do wszystkich rzeczywiście pozytywnych przypadków (TP + FN).
$$
{Recall} = \frac{TP}{TP + FN}
$$
<br>
3. Precyzja (Precision): Proporcja prawidłowo zidentyfikowanych pozytywnych przypadków (TP) do wszystkich przypadków zaklasyfikowanych jako pozytywne przez model (TP + FP).
$$
{Precision} = \frac{TP}{TP + FP}
$$
<br>
4. F1-score - to średnia harmoniczna pomiędzy precyzją (precision) i czułością (recall). Im bliższa jest jedynki, tym lepiej to świadczy o algorytmie klasyfikującym. W najlepszym przypadku przyjmuje wartość 1, kiedy mamy do czynienia z idealną czułością i precyzją.
$$
{F1-score}= \frac{2*TP}{2*TP + FP + FN}
$$

#### Metoda GirdSearch optymalizacji parametrow
Metoda GridSearchCV (Grid Search Cross-Validation) to zaawansowane narzędzie używane w uczeniu maszynowym do optymalizacji hiperparametrów modeli klasyfikacyjnych. Pozwala ona na systematyczne testowanie i porównywanie różnych kombinacji parametrów, aby znaleźć te, które najlepiej radzą sobie z zadaniem klasyfikacji. Jest to niezwykle przydatne w projektach, takich jak klasyfikacja jakości wody, gdzie precyzja modelu ma bezpośrednie przełożenie na zdrowie i bezpieczeństwo ludzi.

### Zasada działania GridSearchCV

GridSearchCV przeprowadza walidację krzyżową dla każdej kombinacji parametrów w zdefiniowanej "siatce" parametrów. Oto główne etapy pracy z GridSearchCV:

1. **Definicja przestrzeni parametrów**:
   Użytkownik określa zbiór potencjalnych wartości dla różnych hiperparametrów. Na przykład, dla modelu SVM można badać różne wartości parametru regularyzacji `C`, różne jądra (`kernel`) takie jak `linear`, `rbf`, a także różne wartości `gamma`.

2. **Walidacja krzyżowa**:
   Dla każdej unikalnej kombinacji parametrów GridSearchCV trenuje model na różnych podzbiorach danych treningowych i ocenia go na pozostałej części zbioru (walidacyjnym). Proces ten jest powtarzany wielokrotnie (zgodnie z ustaloną liczbą fałdów w walidacji krzyżowej), co pozwala na uzyskanie stabilnej oceny wyników modelu dla danej konfiguracji parametrów.

3. **Wybór najlepszego modelu**:
   Po przetestowaniu wszystkich kombinacji, GridSearchCV wybiera zestaw parametrów, który średnio osiągnął najlepsze wyniki w walidacji krzyżowej. Wyniki te mogą być oceniane na podstawie różnych metryk, takich jak dokładność, F1-score czy innych, odpowiednich dla zadania klasyfikacyjnego.



##Metoda K najbliższych sąsiadów
Metoda K-Nearest Neighbors (KNN) jest popularnym algorytmem stosowanym w statystyce i uczeniu maszynowym do klasyfikacji oraz regresji. W klasyfikacji KNN, obiekt jest przypisywany do klasy, do której należy większość jego najbliższych sąsiadów. W regresji KNN, wartość dla nowego punktu jest obliczana jako średnia wartości jego najbliższych sąsiadów. Metoda KNN została szeroko opisana w literaturze naukowej. Jednym z klasycznych źródeł jest praca Covera i Hartsa z 1967 roku, w której przedstawiono podstawy teoretyczne tej metody.

Główną zaletą tej metody jest prostota. Dodatkowo nie wymaga one fazy trenowania modelu, co jest korzystne przy dużych zbiorach danych. Jednakże, warto wziąć pod uwagę, że wydajność może być problematyczna przy dużych zbiorach danych, ponieważ obliczanie odległości jest kosztowne

<b>Podstawowe założenia</b><br>
KNN działa na zasadzie podobieństwa między danymi i zakłada, że obiekty, które są do siebie podobne, znajdują się blisko siebie w przestrzeni cech.

<b>Algorytm KNN</b><br>
1. Wybór liczby K: Na początku należy wybrać liczbę sąsiadów
𝐾, których będziemy brać pod uwagę.
2. Obliczanie odległości: Następnie dla każdego nowego punktu obliczamy odległość do wszystkich punktów w zbiorze uczącym. Najczęściej stosowaną metryką jest odległość euklidesowa:
$$
  𝑑(𝑥,𝑦)={\sqrt{\sum_{i=1}^{n}(x_i - y_i)^2}}
$$
gdzie 𝑥 i 𝑦 to dwa punkty w przestrzeni n-wymiarowej.
3. Wyznaczenie K najbliższych sąsiadów: Spośród wszystkich obliczonych odległości wybieramy K najmniejszych.
4. Klasyfikacja: Punkt zostaje przypisany do klasy, która najczęściej występuje wśród K najbliższych sąsiadów.

##Metoda SVM (Support Vector Machine)
Support Vector Machine (SVM) to nadzorowany algorytm uczenia maszynowego używany głównie do klasyfikacji, ale także do regresji. SVM działa poprzez znalezienie hiperplanu, który najlepiej rozdziela dane na różne klasy w przestrzeni cech.

<b>Podstawowe Założenia</b> <br>
SVM opiera się na koncepcji znalezienia hiperpłaszczyzny o maksymalnym marginesie, czyli odległości między najbliższymi punktami różnych klas. Punkty te, które leżą najbliżej hiperpłaszczyzny, nazywane są wektorami nośnymi (support vectors), ponieważ one wpływają na lokalizację i orientację hiperpłaszczyzny.

Główne założenia metody SVM oparte są na pracy naukowej "Support-Vector Networks" autorstwa Cortesa i Vapnika z 1995 roku, która opisuje teoretyczne podstawy tej metody. Ponadto, różne jądra (kernel functions) używane w SVM, takie jak liniowe, wielomianowe i radialne, zostały szeroko opisane w literaturze naukowej, jak np. w artykule "A tutorial on support vector regression" autorstwa Smolki i Schölkopfa z 2004 roku.

<b>Algorytm SVM:</b>
1. Wybór jądra i parametrów: Na początku należy wybrać odpowiednie jądro i dostosować jego parametry.
2. Optymalizacja marginesu: Następnie algorytm stara się znaleźć hiperpłaszczyznę, która maksymalizuje margines, minimalizując jednocześnie błąd klasyfikacji.
3. Klasyfikacja: Po optymalizacji hiperpłaszczyzny, nowe punkty są klasyfikowane na podstawie tego, po której stronie hiperpłaszczyzny się znajdują.

<b>Hiperplan Separacyjny</b> <br>
W przestrzeni dwuwymiarowej hiperplanem jest linia, która dzieli dane na dwie klasy. W przestrzeni wielowymiarowej jest to płaszczyzna o wyższych wymiarach. $$w⋅x+b=0$$
gdzie w to wektor wag, x to wektor cech, a b to wyraz wolny.

<b>Maksymalizacja Marginesu</b> <br>
SVM stara się znaleźć taki hiperplan, który maksymalizuje margines, czyli odległość między najbliższymi punktami danych każdej klasy a hiperplanem.$$\text{Margines} = \frac{2}{\|w\|}$$

<b>Funkcja Celu</b> <br>
Maksymalizację marginesu można zapisać jako problem minimalizacji następującej funkcji celu:
$$ \min \frac{1}{2} \|w\|^2$$ pod warunkiem, że dla wszystkich punktów treningowych $(x_i,y_i)$:
$$y_i (w \cdot x_i + b) \geq 1$$


</br></br>
Metoda SVM jest ceniona za swoją skuteczność w klasyfikacji danych, zwłaszcza w przypadku danych wysokowymiarowych, oraz za możliwość pracy z różnymi rodzajami danych dzięki różnym jądrom. Jednakże, jej wydajność może być problematyczna przy dużych zbiorach danych, ponieważ czas uczenia może być długi.

## Naiwny klasyfikator Bayesa (Naive Bayes)
Metoda Naiwnych Klasyfikatorów Bayesa (Naive Bayes) jest prostym, ale skutecznym algorytmem uczenia maszynowego stosowanym głównie do klasyfikacji tekstu, takich jak filtracja spamu, analiza sentymentu czy kategoryzacja dokumentów.

Naive Bayes opiera się na twierdzeniu Bayesa, zakładając przy tym, że cechy są niezależne (założenie "naiwności").

<b>Podstawowe założenia</b><br>
Naive Bayes zakłada, że wszystkie cechy są niezależne, co w praktyce jest rzadko spotykane, ale algorytm ten działa zaskakująco dobrze w wielu rzeczywistych zastosowaniach

<b>Twierdzenie Bayesa</b><br>
Twierdzenie Bayesa wyraża się wzorem:
$$P(A∣B) = \frac{P(B|A) \cdot P(A)}{P(B)} $$
​gdzie:
- $P(A∣B)$ to prawdopodobieństwo zdarzenia A przy założeniu, że zdarzenie B zaszło,
- $P(B∣A)$ to prawdopodobieństwo zdarzenia B przy założeniu, że zdarzenie A zaszło,
- $P(A)$ to prawdopodobieństwo zdarzenia A,
- $P(B)$ to prawdopodobieństwo zdarzenia B.
<br><br>

<b>Algorytm Naive Bayes</b><br>
1. Obliczanie prawdopodobieństw apriorycznych:
Dla każdej klasy C obliczamy prawdopodobieństwo P(C).

2. Obliczanie prawdopodobieństw warunkowych:
Dla każdej cechy $x_i$​ obliczamy prawdopodobieństwo $P(x_i|C)$.

3. Zastosowanie twierdzenia Bayesa:
Obliczamy prawdopodobieństwo przynależności do klasy dla nowego punktu danych $x = (x_1, x_2, \ldots, x_n)$ :
$$P(C|x) = \frac{P(C) \cdot P(x_1|C) \cdot P(x_2|C) \cdots P(x_n|C)}{P(x)}$$

4. ​Wybór klasy:
Przypisujemy punkt danych do klasy, dla której prawdopodobieństwo $P(C|x)$ jest największe.

<br><br>
Główne zalety tej metody to prostota, szybkość czy skuteczność przy dużych zbiorach danych. Jednak warto wziąć pod uwaga że algorytm zakłada niezależność cech, co rzadko jest spełnione w praktyce, dlatego może działać słabiej przy cechach silnie zależnych.

## Las Losowy (Random Forest)

Las losowy to zbiór klasyfikatorów, w którym każdy pojedynczy klasyfikator jest drzewem decyzyjnym uczonym do momentu, aż w liściu będą tylko próbki z tej samej klasy. Każdy klasyfikator wchodzący w skład lasu losowego jest uczony na specjalnie wylosowanej dla niego próbce danych, która powstaje poprzez wylosowanie *n* razy ze zwracaniem ze wszystkich *N* próbek uczących. Taka technika generowania danych nazywana jest agregacją bootstrapa.
[Płoński, 2016].

Pierwszy algorytm losowych lasów decyzyjnych został stworzony przez Tin Kam Ho przy użyciu metody losowej podprzestrzeni, która w formule Ho jest sposobem na implementację podejścia "dyskryminacji stochastycznej" do klasyfikacji zaproponowanej przez Eugene'a Kleinberga.

Zaletą tej metody jest zmniejszenie ryzyka nadmiernego dopasowania, które jest typowe dla pojedynczych drzew decyzyjnych, co przekłada się na lepszą skuteczność działania klasyfikatora na nowych danych.

<b>Algorytm </b></br>
1.   Dla każdego drzewa losowo wybierane są próbki ze zbioru danych z powtórzeniami, tworząc zestaw danych treningowych.
2.  Dla każdego drzewa decyzyjnego: </br>
    a) przy każdym podziale węzła losowo wybieranych jest *max_features* cech z dostępnych cech. </br>
    b) dla wybranego podzbioru cech, znajdowany jest najlepszy podział danych maksymalizujący kryterium rozdzialnia, </br>
    c) podział danych w węźle na dwa podzbiory zgodnie z wybraną cechą i wartością progu, powtórzenie procesu podziału dla każdego nowego węzła, aż do spełnienia warunków zatrzymania. </br>
3. Po utworzeniu wszystkich drzew w lesie, każde drzewo przewiduje wynik dla danych wejściowych. W trakcie przygotowania klasy dla próbki jest ona pokazywana wszystkim drzewom decyzyjnym, a jako końcowa decyzja o przynależności do klasy traktowane jest przypisanie do klasy najczęściej wskazywanej przez drzewa decyzyjne w lesie.

##Model Hybrydowy Stacking
Model hybrydowy oparty na technice stacking to zaawansowana metoda klasyfikacji, w której wykorzystuje się połączenie prognoz z różnych modeli bazowych, aby poprawić dokładność i stabilność ostatecznych predykcji. W prezentowanym podejściu użyto dwóch różnych modeli klasyfikacji, RandomForest i SVM (Support Vector Machine), które działają jako modele bazowe, a ich prognozy są następnie wykorzystywane jako wejście dla meta-modelu, którym jest Naiwny Klasyfikator Bayesa (Naive Bayes).

### Opis procesu stosowania modelu hybrydowego:

1. **RandomForestClassifier** - Ten model, działający jako klasyfikator zespołowy, składa się z wielu drzew decyzyjnych. Każde drzewo jest trenowane na losowo wybranych podzbiorach danych, co zapewnia różnorodność w predykcjach i pomaga redukować ryzyko przeuczenia. RandomForest jest wykorzystywany do generowania początkowych prognoz, które są następnie przekazywane jako jedna z cech do meta-modelu.

2. **SVM (Support Vector Machine)** - Ten model jest stosowany ze względu na jego efektywność w radzeniu sobie z wysoko wymiarowymi przestrzeniami cech oraz
zdolność do modelowania nieliniowych granic decyzyjnych. W tym przypadku SVM wykorzystuje jądro RBF (Radial Basis Function) i jest trenowany na tych samych danych co RandomForest, generując niezależne prognozy, które również są przekazywane do meta-modelu.

3. **Meta-model: Naiwny Klasyfikator Bayesa** - W finalnym etapie, Naive Bayes, który jest znany ze swojej szybkości i skuteczności w warunkach niezależności cech, jest używany do syntezowania informacji z prognoz modeli bazowych. W tym celu przyjmuje on wejście złożone z prognoz (prawdopodobieństw) obu modeli bazowych, a następnie generuje końcową predykcję. Stosuje się tutaj metodykę Bayesa z określonymi a priori i parametrem wygładzania variancji, co pozwala na lepsze radzenie sobie z niepewnością i potencjalną niestabilnością danych.

W tym przypadku, poprzez zastosowanie meta-modelu do nauczenia się, jak najlepiej połączyć predykcje z różnych modeli bazowych, można osiągnąć wyższą dokładność niż przy użyciu któ

regokolwiek z modeli bazowych indywidualnie.

## Mierniki
Miernik F1 (f1_score) to jedno z najpowszechniej używanych narzędzi do oceny skuteczności modeli klasyfikacyjnych, szczególnie w przypadku niezrównoważonych danych. Stanowi bardzo ważny pomiar, mogący pomóc zadecydować, który z modeli użyć w swoich badaniach [Frank, 2023].

<b>Podstawowe założenia</b><br>
F1_score opiera się na wskaźnikach precyzji (precision) i czułości (recall)

<b>F1_score</b><br>
F1_score jest zdefiniowany za pomocą następującego wzoru [Frank, 2023]:
$$F1 = 2 \cdot \frac{Precision \cdot Recall}{Precision + Recall} $$
​gdzie:
- $Precision = \frac{TP}{TP + FP}$ określa jak wiele spośród przewidzianych przypadków pozytywnych, rzeczywiście jest pozytywnych,
- $Recall = \frac{TP}{TP + FN}$ określa, jak wiele spośród rzeczywistych przypadków pozytywnych zostało poprawnie przewidziane przez model,
- $TP$ to liczba przypadków, które model poprawnie sklasyfikował jako należące do klasy pozytywnej,
- $TN$ to liczba przypadków, które model poprawnie sklasyfikował jako należące do klasy negatywnej,
- $FP$ to liczba przypadków, które model błędnie sklasyfikował jako należące do klasy pozytywnej
- $FN$ to liczba przypadków, które model błędnie sklasyfikował jako należące do klasy negatywnej

<b>Właściwości</b><br>
- F1_score przyjmuje wartości od 0 do 1, gdzie 1 oznacza najlepszą możliwą skuteczność,
- jest średnią harmoniczną precyzji i czułości, a co za tym idzie jest bardziej czuły na wartości ekstremalne (niska precyzja, czułość)
- ze względu na swój balans, jest szczególnie użyteczny w przypadku, gdy mamy doczynienia z nierównomiernie rozłożymi danymi

<b>Nasze wyniki</b><br>

## Sposób walidacji

Walidacja k-krotna krzyżowa (k-fold corss validation) pozwala na skuteczne oszacowanie ogólnej wydajności modelu. Proces polega na podziale danych na $k$ równych podzbiorów, a następnie na wielokrotnym trenowaniu i testowaniu modelu na różnych zestawach danych. Spośród wielu dostępnych metod walidacji zdecydowaliśmy się na nią, ze względu na rozmiar danych - w tym przypadku walidacja LeaveOneOut okazałaby się zbyt wymagająca obliczeniowo.

<b>Sposób działania</b><br>
- Zbiór jest dzielony losowo na $k$ równych podzbiorów
- W każdej iteracji jeden z $k$ pozdbiorów jest używany jako zbiór testowy, gdzie pozostałe $k-1$ to zbiory treningowe
- Proces powtarzany jest $k$ razy, za każdym razem wybierając inny zbiór jako testowy
- Wyniki każdej iteracji, będące wybranym miernikiem, zapisywane są do jednej tablicy, a następnie obliczana jest ich średnia dająca końcową ocenę modelu

<b>Własności k-krotnej walidacji krzyżowej</b><br>
- Używając wszystkich obserwacji zarówno do treningu, jak i testowania modelu, zapewnia bardziej efektywne wykorzystanie dostępnych danych [Arya, 2022]
- Wykorzystując średnią ze wszystkich wyników jako ocenę końcową, sprawia że pomiar jest mniej podatny na losowe zmiany w danych

<b>Zastosowanie w danych</b><br>
- Aby dodatkowo uchronić się przed losowymi zmianai i nierówną dystrybucją obserwacji reprezentujących obie z klas danych, wykorzystaliśmy podział zbiorów na równe, reprezentatywne dane dzięki funkcji *StratifiedKFolds()*
- Ze względu na rozmiar danych wybraliśmy wartość $k=10$, będącą kompromisem pomiędzy podziałem danych na jak największą ilość podzbiorów, a osiągnięciem satysfakcjonującego czasu obliczeń
- Jako miernik będący oceną wybraliśmy *Precision*, będące miarę określającą stosunek przewidzeń prawdziwie pozytywnych, do wszystkich przewidzeń pozytywnych: $Precision = \frac{TP}{TP + FP}$

## Przyklad użycia modeli na stworzonych sztucznie obserwacjach


```python
min_values = X.min().values
max_values = X.max().values
num_new_samples = 10  # Liczba nowych próbek do wygenerowania
new_samples = np.random.rand(num_new_samples, len(X.columns)) * (max_values - min_values) + min_values

new_samples_df = pd.DataFrame(new_samples, columns=X.columns)

# Skalowanie danych za pomocą MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
new_samples_scaled = scaler.transform(new_samples_df)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(
    n_neighbors=9,
    algorithm='auto',
    leaf_size=10,
    metric='minkowski',
    p=2,
    weights='distance'
)

knn_model.fit(X_train, y_train)
y_pred_knn = knn_model.predict(new_samples_scaled)
print("Metoda KNN dla nowo stworzonych danych")
print(y_pred_knn)
```

```python
from sklearn.svm import SVC
best_params_svc = {
    'C': 10,
    'gamma': 'scale',
    'kernel': 'rbf',
    'random_state': 42
}

# Inicjalizacja modelu SVM z najlepszymi parametrami
svm_model = SVC(**best_params_svc)
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(new_samples_scaled)
print("Metoda SVM dla nowo stworzonych danych")
print(y_pred_svm)
```

```python
from sklearn.naive_bayes import GaussianNB
nb_model = GaussianNB(priors=[0.4, 0.6],var_smoothing= 0.3511191734215131)
nb_model.fit(X_train, y_train)

y_pred_nb = nb_model.predict(new_samples_scaled)
print("Metoda Naive Bayes dla nowo stworzonych danych")
print(y_pred_nb)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(max_depth=20, max_features= 'log2', min_samples_leaf = 1, min_samples_split= 5,n_estimators=150)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(new_samples_scaled)
print("Metoda Random Forest dla nowo stworzonych danych")
print(y_pred_rf)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, max_depth=None, min_samples_split=2, min_samples_leaf=1, max_features='log2', random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf_train = rf_model.predict_proba(X_train)[:, 1]  # Predykcje dla danych treningowych
y_pred_rf_test = rf_model.predict_proba(new_samples_scaled)[:, 1]  # Predykcje dla nowych danych

svm_model = SVC(C=10, kernel='rbf', gamma='scale', random_state=42, probability = True)
svm_model.fit(X_train, y_train)
y_pred_svm_train = svm_model.predict_proba(X_train)[:, 1]  # Predykcje dla danych treningowych
y_pred_svm_test = svm_model.predict_proba(new_samples_scaled)[:, 1]  # Predykcje dla nowych danych


X_meta_train = np.column_stack((y_pred_rf_train, y_pred_svm_train))
meta_model = GaussianNB(priors=[0.5, 0.5], var_smoothing=0.01873817422860384)
meta_model.fit(X_meta_train, y_train)  # Użyj odpowiednich etykiet

# Predykcja za pomocą meta-modelu na nowych danych
X_meta_test = np.column_stack((y_pred_rf_test, y_pred_svm_test))
y_pred_meta = meta_model.predict(X_meta_test)

print("Model Hybrydowy Stacking dla nowo stworzonych danych")
print(y_pred_meta)
```
